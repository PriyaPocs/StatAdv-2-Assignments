Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.


Answer(Q1):


```python
import scipy.stats as stats


CI = 0.95
alpha = 1-CI
sample_mean = 50
sample_std = 5
# Sample size is not given. Assuming that it's 20
sample_size = 20

# Calculate the standard error of the mean (SEM)
standard_error = sample_std / (sample_size ** 0.5)

# Calculate the 95% confidence interval
confidence_interval = stats.norm.interval(CI, loc=sample_mean, scale=standard_error)

print("95% Confidence Interval:", confidence_interval)

```

Interpretation of the results:


The 95% confidence interval represents the range within which we can be 95% confident that the true population mean lies. In this case, the calculated confidence interval will be a range of values with a lower and upper bound. For example, if the calculated confidence interval is (47.8, 52.1), it means that we can be 95% confident that the true population mean falls within this range. It does not mean that there is a 95% probability that the true mean lies in this specific interval, as the true mean is fixed and either falls within the interval or does not. The confidence interval is just a measure of our uncertainty about the true mean based on the sample data. The larger the sample size, the narrower the confidence interval and the more precise our estimate of the true population mean.

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.


Answer(Q2):


```python
import numpy as np
from scipy.stats import chisquare

# blue_count, orange_count, green_count, yellow_count, red_count, brown_count observed values not given.
# To perform chi-squre test, taking some sample values 

blue_count, orange_count, green_count, yellow_count, red_count, brown_count = 30,10,20,10,20,10

# Observed counts for each color
observed_counts = np.array([blue_count, orange_count, green_count, yellow_count, red_count, brown_count])

# Expected distribution percentages
expected_percentages = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

# Calculate the expected counts based on the total number of observations
total_observations = np.sum(observed_counts)
expected_counts = total_observations * expected_percentages

print("total_observations = ",total_observations)

print("observed_counts = ",observed_counts)
print("expected_counts = ",expected_counts)


# Perform the chi-square goodness of fit test
chi2_statistic, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Set the significance level
alpha = 0.05

# Compare the p-value with the significance level to make the decision
if p_value <= alpha:
    print("The distribution of colors of M&Ms in the bag does not match the expected distribution.")
else:
    print("The distribution of colors of M&Ms in the bag matches the expected distribution.")
```

Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data:

Interpret the results of the test.


```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
observed_table = np.array([
    [20, 15],  # Outcome 1: Group A, Group B (Observed counts)
    [10, 25],  # Outcome 2: Group A, Group B (Observed counts)
    [15, 20],  # Outcome 3: Group A, Group B (Observed counts)
])

# Perform the chi-square test for independence
chi2_statistic, p_value, dof, expected_table = chi2_contingency(observed_table)

print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)
```

Now, let's interpret the results of the test:

Chi-square Statistic: The chi-square statistic measures the difference between the observed and expected frequencies in the contingency table. It quantifies how far the observed data deviates from the expected data if there were no association between the outcomes and the groups (Group A and Group B).

P-value: The p-value is the probability of obtaining results as extreme as observed, assuming the null hypothesis is true. In this case, the null hypothesis would be that there is no association between the outcomes and the groups.

Now, let's interpret the p-value:

If the p-value is less than or equal to the chosen significance level (commonly set at 0.05), we reject the null hypothesis. It means there is evidence to suggest that there is a significant association between the outcomes and the groups. In other words, the distribution of outcomes is not the same for Group A and Group B.

If the p-value is greater than the significance level, we fail to reject the null hypothesis. It means we do not have enough evidence to conclude that there is a significant association between the outcomes and the groups. In this case, the distribution of outcomes is consistent with being the same for Group A and Group B.

Always consider the context of the study and the research question when interpreting the results of a statistical test.

Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

Answer(Q5):

To calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12, we can use the t-distribution since the population standard deviation is unknown and the sample size is likely to be small (as we don't have the sample size information).

In Python, we can use the scipy.stats library to calculate the confidence interval. The function we'll use is t.interval().

Let's proceed with the calculation:

```python
import scipy.stats as stats

# Sample statistics
sample_mean = 75
sample_stddev = 12

# Sample size (we would need to provide this value if available)
sample_size = None

# Degrees of freedom (for t-distribution)
# If we know the sample size, use sample_size - 1; otherwise, we can use a large value like 1000
degrees_of_freedom = sample_size - 1 if sample_size is not None else 1000

# Confidence level
confidence_level = 0.90

# Calculate the standard error (standard deviation of the sample mean)
standard_error = sample_stddev / (sample_size**0.5) if sample_size is not None else sample_stddev

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom) * standard_error

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"90% Confidence Interval: ({lower_bound}, {upper_bound})")
```

Note: It is essential to provide the sample size (sample_size) to get a more accurate confidence interval. If the sample size is not available, we can set it to None, but keep in mind that the confidence interval will be less precise.

Interpretation:

The 90% confidence interval for the true population mean, based on the sample data, is (lower_bound, upper_bound). This means that we can be 90% confident that the true population mean lies within this range. In this case, with a mean of 75 and a standard deviation of 12, we are 90% confident that the true population mean falls within the interval (lower_bound, upper_bound).

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.


Answer(Q6):


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Degrees of freedom
df = 10

# Define the range of values for the x-axis
x_values = np.linspace(0, 30, 1000)

# Calculate the probability density function (pdf) for the chi-square distribution
pdf_values = stats.chi2.pdf(x_values, df)

# Plot the chi-square distribution
plt.plot(x_values, pdf_values, label=f"df = {df}")

# Shade the area corresponding to a chi-square statistic of 15
x_shaded = np.linspace(15, 30, 1000)
pdf_shaded = stats.chi2.pdf(x_shaded, df)
plt.fill_between(x_shaded, pdf_shaded, color='gray', alpha=0.5, label="Chi-square statistic = 15")

# Label the axes and add a legend
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density Function')
plt.legend()

# Show the plot
plt.show()
```

The above graph is a plot of the chi-square distribution with 10 degrees of freedom, and the shaded area is corresponding to a chi-square statistic of 15. The x-axis represents the chi-square statistic, and the y-axis represents the probability density function (PDF) of the chi-square distribution. The shaded area indicates the probability of obtaining a chi-square statistic of 15 or higher, given the chi-square distribution with 10 degrees of freedom.

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.


Answer(Q7):

To calculate the 99% confidence interval for the true proportion of people in the population who prefer Coke, we can use the z-score since the sample size is large (n > 30). The formula for the confidence interval is:




Let's calculate the confidence interval using Python:


```python
import numpy as np
import scipy.stats as stats

# Sample statistics
sample_size = 1000
preferred_coke_count = 520

# Calculate the sample proportion
sample_proportion = preferred_coke_count / sample_size

# Calculate the critical value for a 99% confidence level
confidence_level = 0.99
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"99% Confidence Interval: ({lower_bound}, {upper_bound})")
```

Interpretation:

The 99% confidence interval for the true proportion of people in the population who prefer Coke, based on the sample data, is approximately (lower_bound, upper_bound). This means that we can be 99% confident that the true proportion of people who prefer Coke lies within this range, given the sample of 1000 people. In this case, the proportion of people who prefer Coke is estimated to be within the interval (lower_bound, upper_bound).

Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

Answer(Q8):

To conduct a chi-square goodness of fit test for the coin flip data and determine if the observed frequencies match the expected frequencies of a fair coin, we need to set up the hypothesis and perform the test. The null hypothesis (H0) is that the coin is fair, and the alternative hypothesis (H1) is that the coin is biased towards tails.

For a chi-square goodness of fit test, we compare the observed frequencies with the expected frequencies under the assumption of a fair coin. If the test statistic is significant at the chosen significance level (0.05 in this case), we reject the null hypothesis, indicating evidence of bias in the coin towards tails.

Here's how we can conduct the test in Python:

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed_tails = 45
observed_heads = 100 - observed_tails

# Expected frequencies under a fair coin (50% tails, 50% heads)
expected_tails = 100 / 2
expected_heads = 100 / 2

# Create the observed and expected frequency arrays
observed_frequencies = np.array([observed_tails, observed_heads])
expected_frequencies = np.array([expected_tails, expected_heads])

# Perform the chi-square goodness of fit test
chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Set the significance level
alpha = 0.05

# Compare the p-value with the significance level to make the decision
if p_value <= alpha:
    print("The observed frequencies do not match the expected frequencies of a fair coin.")
else:
    print("The observed frequencies match the expected frequencies of a fair coin.")

'''In this case, if the p-value is less than or equal to 0.05, we reject the null 
hypothesis and conclude that the coin is biased towards tails.
If the p-value is greater than 0.05, we fail to reject the null hypothesis, 
indicating that there is no evidence of bias, and the coin appears to be fair.'''
```

Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin.






Use a significance level of 0.01.


Answer(Q10):

To conduct a chi-square test for independence, we first need to set up the observed and expected contingency table. The observed table contains the actual counts of individuals falling into each category, while the expected table contains the counts that we would expect if there were no association between chocolate preference and country of origin. We can then use the chi-square test to determine if there is a significant association between the two variables.

The observed contingency table:

|                 | Milk Chocolate | Dark Chocolate | White Chocolate | Total |
|-----------------|----------------|----------------|-----------------|-------|
| U.S. (n = 500)  | 200            | 150            | 150             | 500   |
| U.K. (n = 500)  | 225            | 175            | 100             | 500   |
| Total           | 425            | 325            | 250             | 1000  |

The expected contingency table (assuming no association):

To calculate the expected counts, we use the formula:

Expected count = (Row total * Column total) / Grand total

|                 | Milk Chocolate | Dark Chocolate | White Chocolate | Total |
|-----------------|----------------|----------------|-----------------|-------|
| U.S. (n = 500)  | 212.5          | 162.5          | 125             | 500   |
| U.K. (n = 500)  | 212.5          | 162.5          | 125             | 500   |
| Total           | 425            | 325            | 250             | 1000  |

Now, we can perform the chi-square test using the formula:

Chi-square = Σ [(Observed count - Expected count)^2 / Expected count]

where Σ denotes the sum over all categories.

Let's calculate the chi-square value step by step:

1. For Milk Chocolate in the U.S.:
(200 - 212.5)^2 / 212.5 ≈ 0.0588

2. For Dark Chocolate in the U.S.:
(150 - 162.5)^2 / 162.5 ≈ 0.0769

3. For White Chocolate in the U.S.:
(150 - 125)^2 / 125 ≈ 6.25

4. For Milk Chocolate in the U.K.:
(225 - 212.5)^2 / 212.5 ≈ 0.0588

5. For Dark Chocolate in the U.K.:
(175 - 162.5)^2 / 162.5 ≈ 0.0769

6. For White Chocolate in the U.K.:
(100 - 125)^2 / 125 ≈ 5.00

Now, we sum all these values:

Chi-square ≈ 0.0588 + 0.0769 + 6.25 + 0.0588 + 0.0769 + 5.00 ≈ 11.57

Next, we need to find the degrees of freedom (df) for the chi-square test. The formula for df is:

df = (Number of rows - 1) * (Number of columns - 1) = (2 - 1) * (3 - 1) = 1 * 2 = 2

Now, we can consult the chi-square distribution table or use a statistical tool to find the critical value for the given significance level (α = 0.01) and df = 2. The critical value for α = 0.01 and df = 2 is approximately 9.21.

Since our calculated chi-square value (11.57) is greater than the critical value (9.21), we reject the null hypothesis. Therefore, we conclude that there is a significant association between chocolate preference and country of origin (U.S. vs. U.K.). In other words, the proportion of people who prefer different types of chocolate is not the same in the U.S. and the U.K.

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.


Answer(Q11):

To conduct a hypothesis test to determine if the population mean is significantly different from 70, we will use a one-sample t-test. Since the population standard deviation is unknown, the t-test is appropriate for this scenario. The hypothesis test will have the following null and alternative hypotheses:

Null Hypothesis (H0): The population mean (μ) is equal to 70.
Alternative Hypothesis (Ha): The population mean (μ) is significantly different from 70.

Let's calculate the t-statistic and then compare it with the critical t-value at a significance level of 0.05.

The formula for the t-statistic in a one-sample t-test is:

t = (sample mean - hypothesized population mean) / (sample standard error)

Where the sample standard error (SE) is calculated as:

SE = sample standard deviation / √(sample size)

Given the information:

Sample mean (x̄) = 72
Sample standard deviation (s) = 10
Sample size (n) = 30
Hypothesized population mean (μ) = 70

Now, let's calculate the t-statistic:

SE = 10 / √30 ≈ 1.8257

t = (72 - 70) / 1.8257 ≈ 1.0966

Next, we need to find the critical t-value for a two-tailed test at a significance level of 0.05 with (n - 1) degrees of freedom, where n is the sample size. In this case, df = 30 - 1 = 29.

Using a t-table or a statistical tool, the critical t-value for a two-tailed test at α = 0.05 and df = 29 is approximately ±2.045.

Since the absolute value of our calculated t-statistic (1.0966) is less than the critical t-value (2.045), we fail to reject the null hypothesis. In other words, at a significance level of 0.05, there is not enough evidence to conclude that the population mean is significantly different from 70. The data does not provide sufficient evidence to reject the null hypothesis that the population mean is equal to 70.
