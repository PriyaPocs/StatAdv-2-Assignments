Q1: What is the difference between a t-test and a z-test? Provide an example scenario where we would
use each type of test.

Answer(Q1):

Both t-tests and z-tests are statistical hypothesis tests used to make inferences about population parameters based on sample data. However, they differ in their assumptions and applications.

1. Z-Test:
A z-test is used when the sample size is large (typically n > 30) and the population standard deviation is known. It assumes that the data is normally distributed. The test statistic for a z-test is based on the standard normal distribution (z-distribution), and the critical values are taken from the standard normal table.

Example scenario for a z-test:
Suppose we want to test whether the mean weight of apples in a large orchard is significantly different from 150 grams. we collect a random sample of 100 apples and find the sample mean weight to be 155 grams. we also know that the population standard deviation of apple weights is 10 grams. In this case, we can use a z-test to determine if the sample mean of 155 grams significantly differs from the hypothesized mean of 150 grams.

2. T-Test:
A t-test is used when the sample size is small (typically n < 30) or when the population standard deviation is unknown. It also assumes that the data is normally distributed. The test statistic for a t-test follows a t-distribution, which has heavier tails compared to the standard normal distribution. The critical values for t-tests depend on the sample size and the desired significance level.

Example scenario for a t-test:
Suppose we want to compare the average test scores of two groups of students (Group A and Group B) to see if there is a significant difference in their performance. Group A consists of 20 students, and Group B consists of 25 students. we collect the test scores of both groups and find that the data is approximately normally distributed. In this case, we can use a t-test to compare the means of the two groups and determine if there is a statistically significant difference in their average test scores.

In summary, use a z-test when we have a large sample size and know the population standard deviation, and use a t-test when we have a small sample size or when the population standard deviation is unknown. Both tests require the data to be approximately normally distributed. Choosing the right test based on the data characteristics is crucial for accurate statistical inference.

Q2: Differentiate between one-tailed and two-tailed tests.


Answer(Q2):

One-tailed and two-tailed tests are concepts in hypothesis testing, which is a statistical method used to make inferences about population parameters based on sample data. The key difference between the two lies in the directionality of the hypothesis being tested and the critical region used for decision-making.

One-tailed tests are used when the hypothesis specifies a particular direction for the effect or difference.Also known as a directional test, a one-tailed test is used when the research hypothesis specifies the direction of the effect or difference between groups. The critical region is only located on one side of the distribution, either the upper tail or the lower tail, depending on the direction specified in the hypothesis. The test is used to determine if the sample data significantly supports the hypothesis in one specific direction.

Two-tailed tests are used when the hypothesis does not specify a particular direction, and we want to check for a significant effect in either direction.Also known as a non-directional test, a two-tailed test is used when the research hypothesis does not specify the direction of the effect or difference between groups. The critical region is located on both sides of the distribution, encompassing both the upper and lower tails. The test is used to determine if the sample data significantly supports the hypothesis in either direction.


Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Answer(Q3):

Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing, which is a statistical method used to draw conclusions about population parameters based on sample data.

1. Type 1 Error (False Positive):
A Type 1 error occurs when we reject a null hypothesis (H0) that is actually true. In other words, we conclude that there is a significant effect or difference when, in reality, there is no such effect or difference in the population. It is also known as a false positive.

2. Type 2 Error (False Negative):
A Type 2 error occurs when we fail to reject a null hypothesis (H0) that is actually false. In other words, we conclude that there is no significant effect or difference when, in reality, there is a true effect or difference in the population. It is also known as a false negative.

You decide to get tested for COVID-19 based on mild symptoms. There are two errors that could potentially occur:

Type I error (false positive): the test result says you have coronavirus, but you actually don’t.

Type II error (false negative): the test result says you don’t have coronavirus, but you actually do.


In summary:
- Type 1 error (False Positive): Rejecting a true null hypothesis.
- Type 2 error (False Negative): Failing to reject a false null hypothesis.

Both Type 1 and Type 2 errors are inherent in hypothesis testing and represent the trade-off between the risk of making incorrect conclusions. In hypothesis testing, we often set the significance level (alpha) to control the probability of Type 1 errors. However, reducing the risk of Type 1 errors may increase the risk of Type 2 errors, and vice versa. Researchers need to carefully consider these potential errors when interpreting the results of hypothesis tests.

Q4:  Explain Bayes's theorem with an example.

Answer(Q4):

Bayes theorem is also known as the Bayes Rule or Bayes Law. It is used to determine the conditional probability of event A when event B has already happened. The general statement of Bayes’ theorem is “The conditional probability of an event A, given the occurrence of another event B, is equal to the product of the event of B, given A and the probability of A divided by the probability of event B.

P(A|B) = (P(B|A) * P(A)) / P(B)

where,
P(A) and P(B) are the probabilities of events A and B
P(A|B) is the probability of event A when event B happens
P(B|A) is the probability of event B when A happens

Example:

A bag contains 6 red, 4 black balls. Two balls are drawn from it as red followed by black. find the probability of the 2nd ball drawn using Bayes thearom

To find the probability of drawing a black ball (2nd ball) given that the 1st ball drawn was red, we can use Bayes' Theorem.

Let's define the events:
A: Drawing a red ball (1st ball)
B: Drawing a black ball (2nd ball)

We are given:
P(A) = Probability of drawing a red ball = Number of red balls / Total number of balls = 6 / (6 + 4) = 6 / 10 = 0.6

Now, we need to find:
P(B|A) = Probability of drawing a black ball given that the 1st ball drawn was red

Bayes' Theorem states:

P(B|A) = (P(A|B) * P(B)) / P(A)

where:
P(A|B) = Probability of drawing a red ball given that the 2nd ball drawn was black (which is 0 because the 2nd ball is drawn after the 1st ball)
P(B) = Probability of drawing a black ball = Number of black balls / Total number of balls = 4 / (6 + 4) = 4 / 10 = 0.4

Now, let's calculate P(B|A):

P(B|A) = (0 * 0.4) / 0.6
P(B|A) = 0 / 0.6
P(B|A) = 0

The probability of drawing a black ball (2nd ball) given that the 1st ball drawn was red is 0. This means that it is impossible to draw a black ball as the 2nd ball if the 1st ball drawn was red. This result is expected since there are no black balls left in the bag after drawing a red ball first.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Answer(Q5):

A confidence interval is a statistical range of values that is used to estimate an unknown population parameter, such as a population mean or proportion. It provides a measure of the uncertainty or variability associated with the sample estimate. The confidence interval is constructed based on the sample data and a chosen level of confidence, typically expressed as a percentage. For example, a 95% confidence interval means that we are 95% confident that the true population parameter lies within the calculated interval.

The formula to calculate a confidence interval depends on the type of parameter being estimated (e.g., mean, proportion) and the distribution of the data. Here's a step-by-step explanation of how to calculate a confidence interval for a population mean, assuming a normal distribution and known population standard deviation:

1. Gather sample data: Obtain a random sample from the population of interest. Let's assume we have a sample of test scores from a class with a known standard deviation of 5.

2. Compute the sample mean: Calculate the mean of the sample data. Let's assume the sample mean is 75.

3. Determine the confidence level: Decide on the desired confidence level. For example, if we want a 95% confidence interval, the confidence level (CL) is 0.95.

4. Find the critical value: Look up the critical value corresponding to the chosen confidence level and the sample size. For a normal distribution, we use the Z-distribution's critical values. In this example, a 95% confidence level corresponds to a Z-critical value of approximately ±1.96 for a two-tailed test.

5. Calculate the standard error: The standard error (SE) is the standard deviation of the sample mean. It is calculated as the population standard deviation divided by the square root of the sample size. In this example, if the sample size is 100, the standard error would be 5 / √100 = 0.5.

6. Compute the confidence interval: Finally, use the formula to calculate the confidence interval:

   Confidence Interval = Sample Mean ± (Z-critical value * Standard Error)

   Confidence Interval = 75 ± (1.96 * 0.5) = (74.02, 75.98)

Interpretation of the confidence interval:

With a 95% confidence level, we can say that we are 95% confident that the true population mean lies within the range of 74.02 to 75.98. This means that if we were to take many random samples and calculate a 95% confidence interval for each sample, approximately 95% of these intervals would contain the true population mean. The confidence interval provides a range of plausible values for the population mean based on the sample data and the chosen level of confidence.

Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.

Answer(Q6):


Bayes' Theorem is a fundamental concept in probability theory that allows us to update the probability of an event based on new evidence or information. It provides a way to calculate the conditional probability of an event given prior knowledge.

The formula for Bayes' Theorem is as follows:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:
- P(A|B) is the conditional probability of event A occurring given that event B has occurred.
- P(B|A) is the conditional probability of event B occurring given that event A has occurred.
- P(A) is the prior probability of event A occurring (before considering the new evidence).
- P(B) is the prior probability of event B occurring (before considering the new evidence).

Let's go through a sample problem to illustrate how to use Bayes' Theorem:

Sample problem:
Suppose we have a box containing red and green balls. The box has 30% red balls and 70% green balls. We also know that the red balls are twice as likely to be selected for sampling as the green balls. Now, if we randomly pick a ball from the box and it turns out to be red, what is the probability that it was one of the green balls?

Solution:
Let's define the events:
- Event A: Selecting a green ball.
- Event B: Observing a red ball.

We are asked to find P(A|B), the probability of selecting a green ball given that we observed a red ball.

Given information:
- P(A) = 0.70 (Prior probability of selecting a green ball)
- P(B|A) = 0.30 (Conditional probability of observing a red ball given that a green ball was selected)
- P(B|not A) = 0.60 (Conditional probability of observing a red ball given that a red ball was selected)
(Note: P(B|not A) = 1 - P(B|A) because the sum of probabilities for complementary events is 1)

Step 1: Calculate the prior probability of observing a red ball (P(B)):
P(B) = P(A) * P(B|A) + P(not A) * P(B|not A)
P(B) = 0.70 * 0.30 + 0.30 * 0.60
P(B) = 0.21 + 0.18
P(B) = 0.39

Step 2: Apply Bayes' Theorem to find the probability of selecting a green ball given that we observed a red ball:
P(A|B) = (P(B|A) * P(A)) / P(B)
P(A|B) = (0.30 * 0.70) / 0.39
P(A|B) = 0.21 / 0.39
P(A|B) ≈ 0.538

So, the probability of the selected ball being green, given that we observed a red ball, is approximately 0.538 or 53.8%.



Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5. Interpret the results.

```python
import scipy.stats as stats


CI = 0.95
alpha = 1-CI
sample_mean = 50
sample_std = 5
# Sample size is not given. Assuming that it's 20
sample_size = 20

# Calculate the standard error of the mean (SEM)
standard_error = sample_std / (sample_size ** 0.5)

# Calculate the 95% confidence interval
confidence_interval = stats.norm.interval(CI, loc=sample_mean, scale=standard_error)

print("95% Confidence Interval:", confidence_interval)
```

The larger the sample size, the narrower the confidence interval and the more precise our estimate of the true population mean.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?

Answer(Q8):
    
The margin of error (MOE) in a confidence interval is a measure of the uncertainty or precision associated with the estimate of a population parameter, such as a mean or proportion. It represents the maximum amount by which the sample estimate is expected to differ from the true population value. The margin of error is typically expressed as a percentage and is directly related to the level of confidence chosen for the interval.

In a confidence interval, the range is constructed around the sample estimate, and the margin of error is calculated by multiplying the standard error (SE) of the sample estimate by the critical value from the appropriate probability distribution. The formula to calculate the margin of error for a given confidence level (CL) is:

Margin of Error (MOE) = Critical Value * Standard Error (SE)

The standard error accounts for the variability of the sample estimate and decreases with larger sample sizes. As the sample size increases, the standard error decreases, which means the margin of error also decreases.

The relationship between sample size and the margin of error can be understood intuitively. When the sample size is larger, the sample estimate becomes more representative of the population, and there is less sampling variability. Therefore, the interval around the sample estimate can be narrower, resulting in a smaller margin of error.

Example scenario:
Let's consider a scenario where we want to estimate the average height of students in a school. We take two different sample sizes, one small and one large, and construct 95% confidence intervals for each.

Small Sample Size:
Sample Size (n1) = 50 students

Large Sample Size:
Sample Size (n2) = 500 students

Assuming the population standard deviation is unknown, we can use the t-distribution for both sample sizes to calculate the 95% confidence intervals.

For the small sample:
MOE1 = t_critical * (s1 / √n1)

For the large sample:
MOE2 = t_critical * (s2 / √n2)
Using a t-table for a 95% confidence level with 499 degrees of freedom, the t_critical value is approximately 1.965.

As we can see, the margin of error for the large sample (0.56 cm) is significantly smaller than the margin of error for the small sample (2.26 cm). Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population

To calculate the z-score for a data point, you can use the formula:

z = (x - μ) / σ

Where:
- x is the value of the data point (in this case, 75).
- μ is the population mean (in this case, 70).
- σ is the population standard deviation (in this case, 5).
- z is the z-score.

Let's calculate the z-score:

z = (75 - 70) / 5
z = 5 / 5
z = 1

In this context, "standard deviation" is a measure of the variability of data points around the mean.

In summary, a z-score allows us to determine the position of a data point relative to the mean of the population and provides a standardized measure of how far the data point deviates from the mean in terms of the population's standard deviation. A positive z-score indicates that the data point is above the mean, while a negative z-score would indicate that the data point is below the mean. The magnitude of the z-score indicates how far the data point is from the mean in terms of standard deviations.


Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average

Step 1: Set up the hypotheses:

H0: μ = 0 (population mean weight loss is zero)

H1: μ ≠ 0 (population mean weight loss is not zero)


Step 2: Calculate the t-statistic:

The t-statistic formula for testing a population mean with an unknown population standard deviation is given by:

t = (x̄ - μ) / (s / √n)

Where:

x̄ is the sample mean (average weight loss in the sample),

μ is the hypothesized population mean (in this case, it is 0),

s is the sample standard deviation (2.5 pounds), and

n is the sample size (50).

t = (6 - 0) / (2.5 / √50)

t = 6 / (2.5 / 7.071)

t ≈ 6 / 0.353

t ≈ 16.99


Step 3: Find the critical t-value:

Since the sample size is 50, we have (50 - 1) degrees of freedom for a two-tailed test at a 95% confidence level. Using a t-table or a statistical software, the critical t-value is approximately ±2.009.

Step 4: Compare the t-statistic with the critical t-value:

Since the t-statistic (16.99) is much larger than the critical t-value (±2.009), we can reject the null hypothesis.

Step 5: Interpret the results:

Based on the results of the hypothesis test, we have enough evidence to conclude that the new weight loss drug is significantly effective at a 95% confidence level. The average weight loss in the sample of 50 participants is significantly different from zero, suggesting that the drug has an impact on weight loss.
